```python
## Import the libraries numpy,pandas,sklearn

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# load the data set
diabetes_data=pd.read_csv('E:\Applied AI related materials\Python notebooks\diabetes.csv')
```

```python
diabetes_data.head(10) ## it gives the first 10 rows of the dataset
```

```python
diabetes_data.tail(10) ## it gives the last 10 rows of the data set
```

```python
diabetes_data.describe()   ##desribe the data and know about the statistics of data like mean, std etc.
```

```python
diabetes_data['Outcome'].value_counts() ##counts the values of different datapoints in outcome column
```

```python
X=diabetes_data.drop(columns='Outcome', axis=1)   ## assign the features to the variabels
Y=diabetes_data['Outcome']
```

```python
scaler=StandardScaler()                                           ##Call StandardScaler() function and assign to varaible
scaler.fit(X)                                                     ## fit the datset X to scaler function
standardized_data=scaler.transform(X)                              ## transform the datset X and standardized the dataset
```

```python
X= standardized_data                   ## assign standardized data to variable X
```

```python
##split the data in train and test data where 20 percent data reserve for test and randomize the data in terms of Y 

X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
## training the  SVM model 

classifier=svm.SVC(kernel='linear')
classifier.fit(X,Y)
```

```python
## Now we check the accuracy of the model using train data

X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('The accuracy of the model on training data is: ',training_data_accuracy)
```

```python
## Now we check the accuracy of the model using test data

X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('The accuracy of the model on test data is: ',test_data_accuracy)
```
